```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import Ridge
```

```python
df = pd.read_csv("C:/Users/RESHMA KEESARI/Downloads/archive (1)/Housing.csv")
df
```

```python
duplicates= df.duplicated().sum()
duplicates
```

```python
x_data = df['area']
y_data = df['bedrooms']
plt.scatter(x_data,y_data)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df['area']=le.fit_transform(df['area'])
df['bedrooms']=le.fit_transform(df['bedrooms'])
df['bathrooms']=le.fit_transform(df['bathrooms'])
df['mainroad']=le.fit_transform(df['mainroad'])
df['basement']=le.fit_transform(df['basement'])
df['parking']=le.fit_transform(df['parking'])
```

```python
x= df[['area','bedrooms','bathrooms','mainroad','basement','parking']]
y= df[['price']]
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=42)
features = df[['area','bedrooms','bathrooms','mainroad','basement','parking']]
target= df[['price']]
```

```python
data_new = pd.concat([features,target], axis=1)
corr_matrix= data_new.corr()
corr_matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, annot= True, cmap='viridis', fmt=".2f")
plt.title("heatmap:features v/s target")
plt.show()
```

```python
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
model.score(x_train,y_train)
```

```python
mse= mean_squared_error(y_test, y_pred)
r2= r2_score(y_test,y_pred)
```

```python
print(f"mean squared error:{mse:.2f}")
print(f"R - squared:{r2:.2f}")
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("Actual prices v/s Predicted prices")
plt.show()
```

```python
# Define new data points for prediction
new_data = [[2555, 3, 2, 1, 0, 2]]

# Predict the prices
predicted_prices = model.predict(new_data)

# Print the results
for i, price in enumerate(predicted_prices):
    print(f"Predicted price for house {i+1}: {price}")
```
